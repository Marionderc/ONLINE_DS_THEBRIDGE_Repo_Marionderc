```python
# Cargamos el dataset cambiando la fila que tiene los nombres de las columnas

df = pd.read_excel('./turismo_2019_2024.xlsx', header = 8)

df
```

```python
# Reseteamos el índice del DataFrame después de cambiar las filas

df.reset_index(drop=True, inplace=True)
```

```python
# viendo el excel, voy a eliminar también las ultimas filas del dataframe que no me sirven:

df = df.iloc[:-12] 

df
```

```python
# voy a eliminar la primera fila que son los totales y no los quiero

df = df.iloc[1:]

df
```

```python
# elegimos las comunidades autonomas junto con las columnas del total de turistas para nuestro primer dataframe:


# Lista de índices de filas que quiero seleccionar
indices_filas_comunidades = [51, 14, 62, 21, 60, 10, 39, 23, 66, 48, 56, 33, 44, 16, 1, 58, 18]

# Creo el primer DataFrame seleccionando las filas y columnas específicas
df_com_total = df.iloc[indices_filas_comunidades, 0:62]  
```

```python
df_com_total
```

```python
# Creamos una copia del DataFrame original para asegurarnos de que no se modifique
df_copy = df.copy()

# Uso drop para eliminar las filas en indices_filas_comunidades de la copia y selecciono las columnas para el segundo dataframe que son las provincias con el total de turistas
df_prov_total = df_copy.drop(indices_filas_comunidades).iloc[:, :62]
```

```python
df_prov_total
```

```python
# Creamos el tercer DataFrame seleccionando las filas en indices_filas_comunidades y las columnas de los turistas españoles


df_com_resi = pd.concat([df.iloc[indices_filas_comunidades, 0], df.iloc[indices_filas_comunidades, 62:123]], axis=1)
df_com_resi
```

```python
# Elimino las filas seleccionadas en la copia y selecciono las columnas de los turistas españoles junto con las provincias para mi cuarto dataframe

df_prov_resi = pd.concat([df_copy.drop(indices_filas_comunidades).iloc[:, 0], df_copy.drop(indices_filas_comunidades).iloc[:, 62:123]], axis=1)
df_prov_resi
```

```python
# Creamos el quinto DataFrame seleccionando las filas de indices_filas_comunidades que son las comunidades y las columnas de extranjeros
df_com_ext = pd.concat([df.iloc[indices_filas_comunidades, 0], df.iloc[indices_filas_comunidades, 123:]], axis=1)
df_com_ext
```

```python
# creamos el sexto y ultimo dataframe de las provincias con los turistas extranjeros. eliminamos las filas seleccionadas en la copia y seleccionamos la primera columna y de la 123 hasta el final
df_prov_ext = pd.concat([df_copy.drop(indices_filas_comunidades).iloc[:, 0], df_copy.drop(indices_filas_comunidades).iloc[:, 123:]], axis=1)
df_prov_ext
```
